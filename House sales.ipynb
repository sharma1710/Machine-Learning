# Splitting the data set

```python
x = df [['bedrooms', 'bathrooms', 'sqft_living', 'floors', 'waterfront', 'grade','sqft_above', 'sqft_basement', 'lat', 'sqft_living15']]
y = df[['price']]
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test= train_test_split(x,y,test_size=0.4)
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
x_test = sc.transform(x_test)
```

# Training and testing the model

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, Y_train)
```

```python
print(reg.intercept_)
```

```python
predictions=reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
import numpy as np
```

```python
print('R-square score:',r2_score(y_test,predictions))
print(' Root Mean squared error:',mean_squared_error(y_test,predictions))
print('mean absolute error:',mean_absolute_error(y_test,predictions))
```
